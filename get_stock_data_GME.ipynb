```python

# %% Libraries
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python

# %% Parameters
end_date = datetime.now().strftime('%Y-%m-%d') # Today
start_date = '2021-01-01'
end_date = '2021-12-31'
interval = '1d'
```

```python

# %% Getting stock market data
# GME
gme = yf.Ticker("gme")

gme_hist = gme.history(start=start_date, end=end_date, interval=interval)
print(gme_hist)
```

```python

# %% Plotting price
plt.figure(figsize=(12, 6))
sns.lineplot(data=gme_hist, x=gme_hist.index, y='Close', label='Closing price', color='b')
plt.xlabel('Date')
plt.title('GME closing price')
plt.legend()
plt.show()
```

```python

# %% Save GME
gme_hist.to_csv("GME_prices.csv")
```
